```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = conv_dates
train1 = data.drop(['id', 'price'],axis=1)
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 400, max_depth = 5, min_samples_split = 2,
          learning_rate = 0.1, loss = 'squared_error')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
n_estimators = clf.get_params()['n_estimators']
t_sc = np.zeros((n_estimators), dtype=np.float64)
```

```python
y_pred = reg.predict(x_test)
```

```python
for i,y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i]=clf.loss_(y_test,y_pred)
```

```python
# Create a dictionary containing the parameters of the model.
params = {"n_estimators": 400}

# Get the number of estimators from the parameters dictionary.
n_estimators = params["n_estimators"]

# Create an array of zeros with the same length as the number of estimators.
t_sc = np.zeros((n_estimators), dtype=np.float64)

# Predict the test data using the regression model.
y_pred = reg.predict(x_test)

# Calculate the loss for each stage of the prediction.
for i, y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i] = clf.loss_(y_test, y_pred)

# Create an array of test scores.
testsc = np.arange((params['n_estimators'])) + 1
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(testsc,clf.train_score_,'b-',label= 'Set dev train')
plt.plot(testsc,t_sc,'r-',label = 'set dev test')
```
